### **NLTK CODE**

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
# Load the required NLTK corpora
nltk.download("punkt")
nltk.download("wordnet")

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
menu = {
    "pizza": ["margherita", "pepperoni", "vegetarian"],
    "spaghetti": ["marinara", "carbonara"],
    "salad": ["caesar", "greek"],
    "chicken": ["grilled", "fried"],
    "cheesecake": ["strawberry", "blueberry"]
}
menu_items = []
for categories, items in menu.items():
    menu_items.extend(items)
print(menu_items)
```

```python
# Function to process an order
def process_order(order):
    # Tokenize the order into individual words
    tokens = word_tokenize(order)

    # Lemmatize the tokens
    lemmas = [lemmatizer.lemmatize(token) for token in tokens]

    # Check if each lemma is a menu item
    items = []
    for lemma in lemmas:
        if lemma in menu_items:
            items.append(lemma)

    return items
```

```python
# Load the customer orders dataset
customer_orders = [
    "I would like a large margherita pizza and a greek salad on the side",
    "Can I get a bowl of carbonara spaghetti and a slice of blueberry cheesecake?",
    "I'd like a grilled chicken sandwich and a caesar salad please",
    "I want a pepperoni pizza and a strawberry cheesecake for dessert"
]
```

```python
for order in customer_orders:
    items = process_order(order)
    print(f"Order: {order}")
    print(f"Items: {items}")
    print("")
```

```python
print("Welcome to our restaurant!")

while True:
    # Get the order from the customer
    order = input("What would you like to order today? ")

    # Process the order
    items = process_order(order)

    if items is None:
      order = input("Sorry this item is not present.")
      while True:
        another_order = input("Would you like to place another order? (yes/no)")
        if another_order.lower() == "yes":
          items = process_order(order)
        else:
          break
    else:

      while True:
        another_order = input("Would you like to place another order? (yes/no)")
        if another_order.lower() == "yes":
          items = process_order(order)
        else:
          break
    break
if items is None:
  print("you ordered nothing.")
else:;
  print(f"You ordered: {items}")
print("Thank you for visiting our restaurant! Have a great day!")
```
